```python
#crop and save

#data folder delle immagini
src_img = "DATA"
src_txt = "new_labels"
output_dir = "cropped"
name_list = os.listdir(src_txt)

#for i in l:
#    name_list.append(i[:5])

for n in name_list:

    name = n[:5]

    bound_txt = os.path.join(src_txt, name + ".txt")
    #print(bound_txt) 
    
    img_path = os.path.join(src_img, name + ".jpg")
    #print(img_path)

    file = open(bound_txt, "r")

    values = []
    values = file.readlines()
    # Rimuove il carattere \n da ogni riga
    v = [riga.strip() for riga in values]

    #print(v)

    # Read the image
    image = cv2.imread(img_path)

    # Extract image dimensions
    height, width, _ = image.shape
    

    #print("file", bound_txt)
    print("image width", width)
    print("image height", height)

    for crop in range(len(v)):
        
        bb = v[crop].split()
        print(bb[1:])

        # Read YOLO format image coordinates
        x_center = float(bb[1])  # Relative x-coordinate of the top-left corner of the region to crop
        y_center = float(bb[2])  # Relative y-coordinate of the top-left corner of the region to crop
        w = float(bb[3])  # Relative width of the region to crop
        h = float(bb[4])  # Relative height of the region to crop

        x_center = round(x_center * width)
        y_center = round(y_center * height)
        w = round(w * width)
        h = round(h * height)
        x = round(x_center - w / 2)
        y = round(y_center - h / 2)
        

        print("bbox width", w)
        print("bbox height", h)

        #values_x= [5*x/100, 7*x/100, 10*x/100]
        #values_y= [5*y/100, 7*y/100, 10*y/100]

        #values_x= [20*x/100, 25*x/100, 30*x/100, 35*x/100, 40*x/100, 45*x/100] 
        #values_y= [20*y/100, 25*y/100, 30*y/100, 35*y/100, 40*y/100, 45*y/100]

        values_x= [40*x/100, 45*x/100, 50*x/100, 60*x/100, 65*x/100, 70*x/100, 80*x/100] 
        values_y= [40*y/100, 45*y/100, 50*y/100, 60*y/100, 65*y/100, 70*y/100, 80*y/100]

        left = round(random.choice(values_x))
        up = round(random.choice(values_y))
        right = round(random.choice(values_x))
        down = round(random.choice(values_y))

        # Crop the image using the calculated pixel values
        cropped_image = image[y-up:y+h+down, x-left:x+w+right]

        # Reformat YOLO label
        new_h = h + down + up
        new_w = w + left + right

        print("new bbox width", new_w)
        print("new bbox height", new_h)

        #resize_w = new_w/width
        #resize_h = new_h/height

        #print("fattore di scala width", resize_w)
        #print("fattore di scala heoght", resize_h)

        final_x_center = (w + 2*left) / (2*new_w)
        final_y_center = (h + 2*up) / (2*new_h)
        final_w = float(bb[3]) / new_w
        final_h = float(bb[4]) / new_h
        #final_w = resize_w
        #final_h = resize_h

        # Save the cropped image
        nameFile_img = name + "_" + str(crop) + ".jpg"
        
        imageClass = int(bb[0])
        #print(imageClass)
        #print(classes[imageClass])

        save_path_img = os.path.join(output_dir, nameFile_img)
        cv2.imwrite(save_path_img, cropped_image)

        # crea nuovo txt con il bounding box del crop
        nameFile_txt = name + "_" + str(crop) + ".txt"

        save_path_txt = os.path.join(output_dir, nameFile_txt)
        new_txt = open(save_path_txt, "w")

        string_value = str(imageClass) + ' ' + str(final_x_center)  + ' ' + str(final_y_center) + ' ' + str(final_w) + ' ' + str(final_h)
        print(string_value)
        new_txt.write(string_value)
        new_txt.close()
      
    file.close()
  
```

```python
src_img = "DATA"
src_txt = "new_labels"
output_dir = "cropped"
name_list = os.listdir(src_txt)

for n in name_list:

    name = n[:5]
    
    image_path = os.path.join(src_img, name + '.jpg')
    bound_txt = os.path.join(src_txt, name + ".txt")

    print(image_path)
    print(bound_txt)

    # reading images
    Image1 = cv2.imread(image_path)
    size = Image1.shape
    
    img_width = size[1]
    img_height = size[0]

    print("img_height", img_height)
    print("img_width", img_width)


    #leggi coordinate da txt

    file = open(bound_txt, "r")

    values = []
    values = file.readlines()
    # Rimuove il carattere \n da ogni riga
    rows = [v.strip() for v in values]

    for row in range(len(rows)):
        
        elem = rows[row].split(" ")
        
        x_center = (float(elem[1]) * img_width)
        y_center = (float(elem[2]) * img_height)
        w = round(float(elem[3]) * img_width)
        h = round(float(elem[4]) * img_height)

        yolo_box = [x_center, y_center, w, h]
        print("yolo box", yolo_box)

        #conversion of bbox coordinates

        x_min = round(x_center - w / 2)
        y_min = round(y_center - h / 2)

        x_max = round(x_center + w / 2)
        y_max = round(y_center + h / 2)

        new_box = [x_min, y_min, x_max, y_max]
        print("new box", new_box)


        # increase the size of the cropped image based on the bbox of the traffic sign
        values_x= [w/2 + 0.3*w/100, w/2 + 0.5*w/100, w/2 + 0.7*w/100, w/2 + 1*w/100, w + 0.3*w/100, w + 0.5*w/100, w + 0.7*w/100]
        values_y= [h/2 + 0.3*h/100, h/2 + 0.5*h/100, h/2 + 0.7*h/100, h/2 + 1*h/100, h + 0.3*h/100, h + 0.5*h/100, h + 0.7*h/100]

        left = round(random.choice(values_x))
        up = round(random.choice(values_y))
        right = round(random.choice(values_x))
        down = round(random.choice(values_y))

        print(left, right, up, down)

        # check if there are not bad values
        crop_x_min = (x_min - left) if (x_min - left) >= 0 else 0
        crop_y_min = (y_min - up) if (y_min - up) >= 0 else 0
        crop_x_max = (x_max + right) if (x_max + right) <= img_width else img_width
        crop_y_max = (y_max + right) if (y_max + right) <= img_height else img_height

        #size of the crop
        crop_w = crop_x_max - crop_x_min
        print("crop_w", crop_w)
        crop_h = crop_y_max - crop_y_min
        print("crop_h", crop_h)

        crop_box = [crop_x_min, crop_y_min, crop_x_max, crop_y_max]
        print("crop box", crop_box)

        """
        xmin_save = left
        ymin_save = up
        xmax_save = crop_w - right
        ymax_save = crop_h - down

        crop_save = str(xmin_save) + ' ' + str(ymin_save) + ' ' + str(xmax_save) + ' ' + str(ymax_save)
        print("crop_save", crop_save)
        """

        # reconversion in yolo format
        new_x_center_yolo = (x_center - crop_x_min) / crop_w
        new_y_center_yolo = (y_center - crop_y_min) / crop_h
        new_w_yolo = w / crop_w
        new_h_yolo = h / crop_h

        yolo_format = [new_x_center_yolo, new_y_center_yolo, new_w_yolo, new_h_yolo]
        print("new yolo format", yolo_format)

        cropped_image = Image1[crop_y_min: crop_y_min + crop_h, crop_x_min:crop_x_min + crop_w]

        # Save the cropped image
        nameFile_img = name + "_" + str(row) + ".jpg"
        
        imageClass = int(elem[0])
        #print(imageClass)
        #print(classes[imageClass])

        save_path_img = os.path.join(output_dir, nameFile_img)
        cv2.imwrite(save_path_img, cropped_image)

        # crea nuovo txt con il bounding box del crop
        nameFile_txt = name + "_" + str(row) + ".txt"

        save_path_txt = os.path.join(output_dir, nameFile_txt)
        new_txt = open(save_path_txt, "w")

        string_value = str(imageClass) + ' ' + str(new_x_center_yolo)  + ' ' + str(new_y_center_yolo) + ' ' + str(new_w_yolo) + ' ' + str(new_h_yolo)
        print(string_value)
        new_txt.write(string_value)
        new_txt.close()

        

    file.close()
```
